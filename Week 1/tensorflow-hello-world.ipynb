```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

lets model a function where the y is 1 less than twice of x.
y = 2x-1

EX:
x = 2, -1, 3, 4, 0
y = 3, -3, 5, 7, -1

```python
#model with simplest unit i.e just 1 neuron and input shape of 1, just 1 dense layer
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# check the model summary
model.summary()
```

```python
# compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# lets create the data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 3.0, 7.0], dtype=float)
```

```python
# fir the model on train data
model.fit(xs, ys, epochs=500)
```

```python
# test model on test data
xtest = np.array([10.0, 5.0, 12.0], dtype=float)
print(model.predict(xtest))
```

the result we expected is 19.0, 9.0, 23.0, model prediction is close to right answer but with deviation which is due the amount of data we provided as it is not enough for the model
